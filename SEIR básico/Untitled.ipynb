```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def modelo_seir(val_inic, param, t):
    S_0, E_0, I_0, R_0 = val_inic
    S, E, I, R = [S_0], [E_0], [I_0], [R_0]
    alpha, beta, gamma = param
    dt = t[1] - t[0]
    for _ in t[1:]:
        next_S = S[-1] - (beta*S[-1]*I[-1])*dt
        next_E = E[-1] + (beta*S[-1]*I[-1] - alpha*E[-1])*dt
        next_I = I[-1] + (alpha*E[-1] - gamma*I[-1])*dt
        next_R = R[-1] + (gamma*I[-1])*dt
        S.append(next_S)
        E.append(next_E)
        I.append(next_I)
        R.append(next_R)
    return np.stack([S, E, I, R]).T
```

```python
# Definir Parámetros
t_final = 50
dt = .1
t = np.linspace(0, t_final, int(t_final/dt) + 1)
N = 10000
val_inic = 1 - 1/N, 1/N, 0, 0
alpha = 0.5
beta = 1.75
gamma = 0.5
param = alpha, beta, gamma
# Correr simulación
resultados = modelo_seir(val_inic, param, t)
```

```python
resultados
```

```python
plt.plot(t,resultados)
plt.legend(["Suceptibles","Expuestos","Infectados","Recuperados"])
```
